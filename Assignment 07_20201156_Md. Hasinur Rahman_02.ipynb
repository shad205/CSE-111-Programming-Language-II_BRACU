```python
#task-2

class Vehicle: 
    def __init__(self): 
        self.x = 0 
        self.y = 0 
    def moveUp(self): 
        self.y+=1
    def moveDown(self): 
        self.y-=1
    def moveRight(self): 
        self.x+=1 
    def moveLeft(self): 
        self.x-=1 
    def __str__(self): 
        return '('+str(self.x)+' , '+str(self.y)+')' 

class  Vehicle2010(Vehicle):
    def moveLowerLeft(self):
        super().moveDown()
        super().moveLeft()
    def moveUpperRight(self):
        super().moveUp()
        super().moveRight()
    def moveLowerRight(self):
        super().moveDown()
        super().moveRight()
    def moveUpperLeft(self): 
        super().moveUp()
        super().moveLeft()
    def equals(self,other):
        self.other=other
        if self.x==other.x and self.y==other.y:
            return 'True'
        else:
            return 'False'


print('Part 1') 
print('------') 
car = Vehicle() 
print(car) 
car.moveUp() 
print(car) 
car.moveLeft() 
print(car) 
car.moveDown() 
print(car) 
car.moveRight() 
print(car) 
print('------') 
print('Part 2') 
print('------') 
car1 = Vehicle2010() 
print(car1) 
car1.moveLowerLeft() 
print(car1) 
car2 = Vehicle2010() 
car2.moveLeft() 
print(car1.equals(car2)) 
car2.moveDown() 
print(car1.equals(car2))
```

```python
#Task 3


class Tournament:
    def __init__(self,name='Default'):
        self.__name = name
    def set_name(self,name):
        self.__name = name
    def get_name(self):
        return self.__name
class Cricket_Tournament(Tournament):
    def __init__(self,*info):
        if len(info)==0:
            self.__name='Default'
            self.__num=0
            self.__type='No type'
        elif len(info)==3:
            self.__name=info[0]
            self.__num=info[1]
            self.__type=info[2]
    def detail(self):
        return f"Cricket Tournament Name: {self.__name}\nNumber of Teams: {self.__num}\nType: {self.__type}"
    
class Tennis_Tournament(Tournament):
    def __init__(self,*dtl):
          if len(dtl)==2:
            self.__name=dtl[0]
            self.__num=dtl[1]
            self.__type=None
    def detail(self):
        return f"Tennis Tournament Name: {self.__name}\nNumber of Teams: {self.__num}"

ct1 = Cricket_Tournament()
print(ct1.detail())
print("-----------------------")
ct2 = Cricket_Tournament("IPL",10,"t20")
print(ct2.detail())
print("-----------------------")
tt = Tennis_Tournament("Roland Garros",128)
print(tt.detail())
```

```python
#task-4


class Product:
    def __init__(self,id, title, price):
        self.__id = id
        self. __title = title
        self. __price = price
    def get_id_title_price(self):
        return "ID: "+str(self.__id)+" Title:"+self.__title+"Price: "+str(self.__price)
class Book(Product):
    def __init__(self,id,name,price,isbn,pub):
        self.__id=id
        self.__title=name
        self.__price=price
        self.__ISBN=isbn
        self.__Publisher=pub
    def printDetail(self):
        return f"ID: {self.__id} Title: {self.__title} Price: {self.__price}\nISBN :{self.__ISBN} Publisher:{self.__Publisher}"
class CD(Product):
    def __init__(self,id,name,price,band,time,genre):
        self.__id=id
        self.__title=name
        self.__price=price
        self.__Band=band
        self.__Duration=time
        self.__Genre=genre

    def printDetail(self):
        return f"ID: {self.__id} Title: {self.__title} Price: {self.__price}\nBand: {self.__Band} Duration:{self.__Duration}minutes\nGenre: {self.__Genre}"


book = Book(1,"The Alchemist ",500,"97806"," HarperCollins")
print(book.printDetail())
print("-----------------------")
cd = CD(2,"Shotto",300,"Warfaze",50,"Hard Rock")
print(cd.printDetail())
```

```python
#Task 6

class Shape:
    def init(self, name='Default', height=0, base=0):
        self.area = 0
        self.name = name
        self.height = height
        self.base = base
    def get_height_base(self):
        return "Height: "+str(self.height)+",Base: "+str(self.base)
    def calcArea(self):
        if self.name=='Trapezoid':
            self.area=float(((self.base+self.side_A)/2)*self.height)
        elif self.name=='Triangle':
            self.area=float(.5*self.base*self.height)
        else:
            self.area=0.0
    def printDetail(self):
        if self.name=='Trapezoid':
            return f'Shape name: {self.name}\nHeight: {self.height}, Base: {self.base}, Side_A: {self.side_A}\nArea: {self.area}'
        else:
            return f'Shape name: {self.name}\nHeight: {self.height}, Base: {self.base}\nArea: {self.area}'
class triangle(Shape):
    def __init__(self, name='Default', height=0, base=0):
        self.area = 0
        self.name = name
        self.height = height
        self.base = base
        
    
class trapezoid(Shape):
    def __init__(self, name, height, base,side_A):
        self.area = 0
        self.name = name
        self.height = height
        self.base = base
        self.side_A=side_A

#write your code here
tri_default = triangle()
tri_default.calcArea()
print(tri_default.printDetail())
print('--------------------------')
tri = triangle('Triangle', 10, 5)
tri.calcArea()
print(tri.printDetail())
print('---------------------------')
trap = trapezoid('Trapezoid', 10, 6, 4)
trap.calcArea()
print(trap.printDetail())
```

```python
#task7


class Football: 
    def __init__(self, team_name, name, role): 
        self.__team = team_name 
        self.__name = name 
        self.role = role 
        self.earning_per_match = 0 
    def get_name_team(self): 
        return 'Name: '+self.__name+', Team Name: ' +self.__team 

    
class Player(Football):
    def __init__(self, team_name, name, role,goal,match):
        self.match=match
        self.goal=goal
        super().__init__(team_name,name, role)
    def calculate_ratio(self):
        return ("Goal Ratio: "+str(self.goal/self.match))
    def print_details(self):
        print(self.get_name_team())
        print("Team Role:",self.role)
        print("Total Goal: "+str(self.goal)+", Total Played: "+str(self.match))
        print(self.calculate_ratio())
        print("Match Earning: "+str((self.goal*1000)+(self.match*10))+"k")
class Manager(Football):
    def __init__(self, team_name, name, role,win):
        self.win=win
        super().__init__(team_name, name, role)
    def print_details(self):
        print(self.get_name_team())
        print("Team Role:",self.role)
        print("Total Win:",self.win)
        print ("Match Earning:",str(self.win*1000)+"k")

#=================================================== 
player_one = Player('Juventus', 'Ronaldo', 'Striker', 25, 32) 
player_one.calculate_ratio() 
player_one.print_details() 
print('------------------------------------------') 
manager_one = Manager('Real Madrid', 'Zidane', 'Manager', 25) 
manager_one.print_details()
```
