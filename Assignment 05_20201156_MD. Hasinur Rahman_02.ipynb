```python
#TASK-1


class Marks:
    def __init__(self, mark):
        self.mark = mark
    def __add__(self, other):
        result = self.mark + other.mark
        return Marks(result) 
    
    
Q1 = Marks(int(input("Quiz 1 (out of 10): ")))
Q2 = Marks(int(input("Quiz 2 (out of 10): ")))
Lab = Marks(int(input("Lab (out of 30): ")))
Mid = Marks(int(input("Mid (out of 20): ")))
Final = Marks(int(input("Final (out of 30): ")))
total = Q1 + Q2 + Lab + Mid + Final
print("Total marks: {}".format(total.mark))
```

```python
#TASK-4


class Color:
    def __init__(self, color):
        self.clr = color
    def __add__(self, other):
        if (self.clr == "red" or self.clr == "blue") and (other.clr == "blue" or other.clr == "red"):
            return Color("Violet")
        elif (self.clr == "red" or self.clr == "yellow") and (other.clr == "yellow" or other.clr == "red"):
            return Color("Orange")
        elif (self.clr == "yellow" or self.clr == "blue") and (other.clr == "blue" or other.clr == "yellow"):
            return Color("Green")
        
        
C1 = Color(input("First Color: ").lower())
C2 = Color(input("Second Color: ").lower())
C3 = C1 + C2
print("Color formed:", C3.clr)
```

```python
#TASK-5


import math
class Circle:
    def __init__(self, radius):
        self.__radius = radius
    def __add__(self, other):
        new_object = self.__radius + other.__radius
        return Circle(new_object)
    def area(self):
        return math.pi * (self.__radius ** 2)
    def getRadius(self):
        return self.__radius
    def setRadius(self, new_radius):
        self.__radius = new_radius
        
        
c1 = Circle(4)
print("First circle radius:" , c1.getRadius())
print("First circle area:" ,c1.area())
c2 = Circle(5)
print("Second circle radius:" ,c2.getRadius())
print("Second circle area:" ,c2.area())
c3 = c1 + c2
print("Third circle radius:" ,c3.getRadius())
print("Third circle area:" ,c3.area())
```

```python
#Task-6


class Triangle:
    def __init__(self, base, height):
        self.__base = base
        self.__height = height
    def __sub__(self, other):
        new_object_base = self.__base - other.__base
        new_object_height = self.__height - other.__height
        return Triangle(new_object_base, new_object_height)
    def area(self):
        return 0.5 * (self.__base * self.__height)
    def getBase(self):
        return self.__base
    def setBase(self, new_base):
        self.__base = new_base
    def getHeight(self):
        return self.__height
    def setHeight(self, new_height):
        self.__height = new_height 
        
        
t1 = Triangle(10, 5)
print("First Triangle Base:" , t1.getBase())
print("First Triangle Height:" , t1.getHeight())
print("First Triangle area:" ,t1.area())
t2 = Triangle(5, 3)
print("Second Triangle Base:" , t2.getBase())
print("Second Triangle Height:" , t2.getHeight())
print("Second Triangle area:" ,t2.area())
t3 = t1 - t2
print("Third Triangle Base:" , t3.getBase())
print("Third Triangle Height:" , t3.getHeight())
print("Third Triangle area:" ,t3.area())
```

```python
#TASK-8


class Coordinates:
    def __init__(self, number1, number2):
        self.number1 = number1
        self.number2 = number2
    def __sub__(self, other):
        objN1 = self.number1 - other.number1
        objN2 = self.number2 - other.number2
        new_obj = Coordinates(objN1, objN2)
        return (new_obj)
    def __mul__(self, other):
        objN1 = self.number1 * other.number1
        objN2 = self.number2 * other.number2
        new_obj = Coordinates(objN1, objN2)
        return (new_obj)
    def __eq__(self, other):
        if (self.number1 == other.number1) and (self.number2 == other.number2):
            return "The calculated coordinates are the same."
        else:
            return "The calculated coordinates are NOT the same."
    def detail(self):
        return ("("+str(self.number1)+","+str(self.number2)+")")
    
    
p1 = Coordinates(int(input()),int(input()))
p2 = Coordinates(int(input()),int(input()))
p4 = p1 - p2
print(p4.detail())
p5 = p1 * p2
print(p5.detail())
point_check = (p4 == p5)
print(point_check)
```
